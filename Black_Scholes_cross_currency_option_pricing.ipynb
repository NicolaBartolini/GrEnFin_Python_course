### Setting

Let $S_{d}(t),S_{f}(t)$ be the stock in the domestic and foreign market respectively, and let $B_{d}(t),B_{f}(t)$ be the risk-free bank accounts in the domestic and foreign market respectively, and let $X(t)$ be the stochastic exchange rate $\frac{domestic}{foreign}$. Let's denote by $\tilde{B}_f(t)=B_f(t)X(t)$ the foreign back account in the domestic economy, and by $\tilde{S}_f(t)=B_f(t)X(t)$ the foreign stock in the domestic economy.

### Assumption 1

Under the objective measure $P$ we assume the following dynamics:

1) $dB_d(t) = r_d B_d(t)dt$
2) $dB_f(t) = r_f B_f(t)dt$
3) $dX(t) = \mu_X X(t)dt + \sigma_X X(t)dW_t^{P}$

Then by Ito's lemma we have the dynamic of $\tilde{B}_f(t)$ under the objective measure:
$$
d\tilde{B}_f(t) = \tilde{B}_f(t)(\mu_X + r_f)dt + \sigma_X \tilde{B}_f(t) dW_t^P
$$

So we see that $\tilde{B}_f(t)$ in the domestic markets behaves like a stock with drift being the sum of the exchange-rate drift foreign risk-free rate. Then, to price a contingent claim on $\tilde{B}_f(t)$ its risk neutral dynamic is:

$$
d\tilde{B}_f(t) = \tilde{B}_f(t)r_ddt + \sigma_X \tilde{B}_f(t) dW_t^Q
$$

Given the risk-neutral dynamic of $\tilde{B}_f(t)$ can be obtained via the Ito's lemma of $X(t) = \frac{\tilde{B}_f(t)}{B_f(t)}$:

$$
dX(t) = X(t)(r_d - r_f)dt + X(t) \sigma_XdW_t^Q
$$
Then, pricing European call and put can be priced via the Black-Scholes formula:
$$
\begin{align}
& C(t,T) = E^Q[e^{-r_d(T-t)}(X(T) - K)^+|\mathcal{F}_t] = X(t)e^{-r_f(T-t)}\Phi(d1) - Ke^{-r_d(T-t)}\Phi(d2) \\
& P(t,T) = E^Q[e^{-r_d(T-t)}(K - X(T))^+|\mathcal{F}_t] = Ke^{-r_d(T-t)}\Phi(-d2) - X(t)e^{-r_f(T-t)}\Phi(-d1),
\end{align}
$$
where
$$
\begin{align}
& d1 = \frac{1}{\sigma_X\sqrt{T-t}}\left\{\log\left(\frac{X(t)}{K} \right) + \left(r_d - r_f + \frac{1}{2}\sigma_X^2 \right)(T-t) \right\} \\
& d2 = d1 - \sigma_X\sqrt{T-t}
\end{align}
$$

#### Remark:
We can see that under the assumption of a market driven by a GBM (Geometric Brownian Motion) the pricing a contingent claim written on $X(t)$ is the same pricing problem of contingent claims on dividend paying stocks

### Assumption 2

Under the objective measure $P$ we assume the following dynamics for the entire economy:

1) $dB_d(t) = r_d B_d(t)dt$
2) $dB_f(t) = r_f B_f(t)dt$
3) $dX(t) = \mu_X X(t)dt + \sigma_X X(t)dW_t^{P}$
4) $dS_d(t) = \mu_d S_d(t)dt + \sigma_d S_d(t)dW_t^{P}$
5) $dS_f(t) = \mu_f S_f(t)dt + \sigma_f S_f(t)dW_t^{P}$

Then the dynamics of $\tilde{B}_f(t)=B_f(t)X(t)$ and $\tilde{S}_f(t)=B_f(t)X(t)$ can be obtained by Ito's lemma:
$$
\begin{align}
& d\tilde{B}_f(t) = \tilde{B}_f(t)( \mu_X + r_f)dt + \sigma_X\tilde{B}_f(t)dW_t^P \\
& d\tilde{S}_f(t) = \tilde{S}_f(t)(\mu_f + \mu_X +\sigma_f\sigma_X^T) + \sigma_f\tilde{S}_f(t)dW_t^P \\
\end{align}
$$
Then, by applying Ito's lemma we get the risk-neutral $Q$-dynamics:
$$
\begin{align}
& dX(t) = X(t)(r_d - r_f)dt + X(t) \sigma_XdW_t^Q \\
& d\tilde{B}_f(t) = \tilde{B}_f(t)r_ddt + \sigma_X \tilde{B}_f(t) dW_t^Q \\
& d\tilde{B}_f(t) = \tilde{B}_f(t)r_d dt + \sigma_X\tilde{B}_f(t)dW_t^Q \\
& d\tilde{S}_f(t) = \tilde{S}_f(t)r_d dt + (\sigma_X + \sigma_f)\tilde{S}_f(t)dW_t^Q \\
& dS_d(t) = r_d S_d(t)dt + \sigma_d S_d(t)dW_t^{Q}
\end{align}
$$
So, now we have more sources of randomness than as in the classic Black-Scholes framework:

1) Foreign call struk in domestic currency option $\Pi(t, X(t), S_f(t)) = E^{Q}[e^{-r_d(T-t)}(X(T)S_f(T)-K_d)^+|\mathcal{F}_t] = E^{Q}[e^{-r_d(T-t)}(\tilde{S}_f-K_d)^+|\mathcal{F}_t]$
2) Foreign put struk in domestic currency option $\Pi(t, X(t), S_f(t)) = E^{Q}[e^{-r_d(T-t)}(K_d - X(T)S_f(T))^+|\mathcal{F}_t] = E^{Q}[e^{-r_d(T-t)}(K_d - \tilde{S}_f)^+|\mathcal{F}_t]$
3) Foreign call, struck in foreign currency $F(S^f_T,T) = (S^f_T -K^f)^+$, then by no arbitrage we have that $\Pi(t,T) = X_t E^{Q^f}[e^{-r_f(T-t)}(S_T^f-K^f)^+|\mathcal{F}_t]$ (So just use the classic Black-Scholes formula readapted for the exchange rate)
4) Foreign put, struck in foreign currency $F(S^f_T,T) = (K^f - S^f_T)^+$, then by no arbitrage we have that $\Pi(t,T) = X_t E^{Q^f}[e^{-r_f(T-t)}(K^f - S_T^f-)^+|\mathcal{F}_t]$ (So just use the classic Black-Scholes formula readapted for the exchange rate)

So, from now on since we have more sources of randomness we will consider a multivariate correlated Brownian Motion

### Explicit formula for the foreign options, struck in domestic currency

#### The European call option
$$
\begin{split}
    & \Pi(t, X(t), S_f(t)) = E^{Q}[e^{-r_d(T-t)}(X(T)S_f(T)-K_d)^+|\mathcal{F}_t] = E^{Q}[e^{-r_d(T-t)}(\tilde{S}_f-K_d)^+|\mathcal{F}_t] \\
    & \Pi(t, X(t), S_f(t)) = X_t S_f \Phi(d_1) - Ke^{-r_d(T-t)} \Phi(d2) \\
    & d1 = \frac{\log\left(\frac{X_t S_t}{K_d} \right) + \left(r_d+\frac{1}{2}||\sigma_f + \sigma_X||^2\right)(T-t)}{\sigma\sqrt{T-t}} \\
    & d2 = d1 - ||\sigma_f + \sigma_X||\sqrt{T-t} \\
    & ||\sigma_f + \sigma_X|| = \sqrt{\sigma_f^2 + \sigma_X^2 + 2\sigma_X\sigma_f\rho}
\end{split}
$$
